# 1. Drawing shapes

### Import libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Create black image to work with

```python
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
black_img.shape
```

### Render the image

```python
plt.imshow(black_img)
```

### Draw a cricle

```python
cv2.circle(img=black_img, center=(400,100), radius=50, color=(255, 0,0), thickness=8)

plt.imshow(black_img)
```

### Filled Circle

```python
cv2.circle(img=black_img,
           center=(400,200),
           radius=50,
           color=(0,255,0),
           thickness=-1)

plt.imshow(black_img)
```

### Draw a rectangle

```python
cv2.rectangle(black_img,
             pt1=(200,200),
             pt2 =(300,300),
             color=(0,255,0),
             thickness=5)

plt.imshow(black_img)
```

### Draw Triangle

```python
vertices = np.array([[10,450],
                    [110, 350],
                    [190, 450]],
                   np.int32)

pts = vertices.reshape(-1,1,2)
cv2.polylines(black_img, [pts], isClosed=True,color=(0,0,255),thickness=5)
plt.imshow(black_img)
```

### Filled Rectangle

```python
cv2.rectangle(black_img,
             pt1=(100,100),
             pt2 =(200,200),
             color=(137,79,213),
             thickness=-1)

plt.imshow(black_img)
```

### Filled Triangle

```python
vertices = np.array([[300,450],
                    [400, 350],
                    [500, 450]],
                   np.int32)

pts = vertices.reshape(-1,1,2)
cv2.fillPoly(black_img,
              [pts],
              color=(255,167,201))
plt.imshow(black_img)
```

### Draw Line

```python
cv2.line(black_img, 
        pt1=(512,0),
        pt2=(0, 512),
        color=(255,0,255),
        thickness=3)

plt.imshow(black_img)
```

### Write Text

```python
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(black_img,
           text='Rhyme',
           org=(210, 500),
           fontFace=font,
           fontScale=3,
           color=(255,255,0),
           thickness=3,
           lineType=cv2.LINE_AA)
plt.imshow(black_img)
```
